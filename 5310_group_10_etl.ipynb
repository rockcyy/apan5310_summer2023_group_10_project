### Ticket table

```python
#combine columns to generate the departure and arrival timestamp
ticket_data = flight_data[['Year','Month','DayofMonth','DepTime','ArrTime','UniqueCarrier','FlightNum']].dropna()

ticket_data['DepTime'] = ticket_data['DepTime'].astype(int)
ticket_data['ArrTime'] = ticket_data['ArrTime'].astype(int)

def format_time(year, month, day, time):
    time_str = str(time).zfill(4)
    return f"{year}-{month:02d}-{day:02d} {time_str[:2]}:{time_str[2:]}"

ticket_data['departure_time'] = ticket_data.apply(lambda row: format_time(row['Year'], row['Month'], row['DayofMonth'], row['DepTime']), axis=1)
ticket_data['arrival_time'] = ticket_data.apply(lambda row: format_time(row['Year'], row['Month'], row['DayofMonth'], row['ArrTime']), axis=1)

def airline_company_id(x):
    if x == 'WN':
        return 1
    else:
        return 2

ticket_data['airline_company'] = ticket_data['UniqueCarrier'].apply(airline_company_id)

#combine columns to generate flight number
ticket_data['flight_number'] = ticket_data['UniqueCarrier'] + ticket_data['FlightNum'].astype(str)

ticket_data = ticket_data[['flight_number','departure_time','arrival_time','airline_company']]

#randomly assign seat number
rows = 50
seats_per_row = ['A', 'B', 'C', 'D', 'E', 'F']
all_seats = [f'{i}{seat}' for i in range(1, rows + 1) for seat in seats_per_row]
random.shuffle(all_seats)
ticket_data['seat_number'] = ticket_data['flight_number'].apply(lambda x: random.choice(all_seats))

#randomly assign ticket price
ticket_data['ticket_price'] = np.around(np.random.uniform(10, 100, ticket_data.shape[0]),2)

#randomly assign departure and arrival address
ticket_data['arrival_airport_address'] = np.random.randint(1, 50, ticket_data.shape[0])
ticket_data['departure_airport_address'] = np.random.randint(1, 50, ticket_data.shape[0])

ticket_data.insert(0, 'ticket_id', range(1, 1 + len(ticket_data)))
ticket_data.to_sql(name='ticket', con=engine, if_exists='append', index=False)
```
